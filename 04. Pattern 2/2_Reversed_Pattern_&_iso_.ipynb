star increases space decrease so we have to create 2 loop of space & star


```python
n = int(input())

i = 1
while i <= n:

  space = 1
  while space <= n-i:
    print(' ', end = '')
    space = space + 1
  
  star = 1
  while star <= i:
    print('*', end = '')
    star = star + 1
  print()
  i = i + 1
```

```python
n = int(input())

i = 1
while i <= n:

# for spaces
  space = 1
  while space <= n-i:
    print(' ', end = '')
    space = space + 1
  
  # increase seq
  p = 1
  j = 1
  while j <= i:
    print(p, end = '')
    p = p + 1
    j = j + 1

  #decrease seq
  p = i-1
  while p >= 1:
    print(p, end = '')
    p = p-1

  print()
  i = i + 1
```
