```python
arr = [1,32,4,0,6,8]

def selectionSort(arr):
  lenght =  len(arr)

# put the correct elemnet at ith position
  for i in range(lenght - 1):
    minIndex = i # this will give 4th value as mid 

# calculating index of min element for this iteration
    for j in range(i+1, lenght):
      if (arr[j] < arr[minIndex]): 
        minIndex = j

      arr[i], arr[minIndex] = arr[minIndex], arr[i]
      
```

```python
arr = [1,32,4,0,6,8]
selectionSort(arr)
print(arr)
```
